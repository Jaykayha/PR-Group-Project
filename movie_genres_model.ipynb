````python
#COMEDY COLUMN ONLY!!!

from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(new_df.poster_path, new_df.action, test_size = 0.2, random_state = 42)

# split test into test and val
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size = 0.5, random_state = 42)

# create new dataframes
data_train = pd.DataFrame.from_dict({"path":X_train, "action":[str(value) for value in y_train]})
data_val = pd.DataFrame.from_dict({"path":X_val, "action":[str(value) for value in y_val]})
data_test = pd.DataFrame.from_dict({"path":X_test, "action":[str(value) for value in y_test]})
````

````python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# put data in format readable by imagenet

batch_size = 64
size    = (224, 224)
data_generator = ImageDataGenerator(
    horizontal_flip = True,
    #rescale = 1./255,
    #preprocessing_function = tf.keras.applications.mobilenet_v2.preprocess_input
)

train_set = data_generator.flow_from_dataframe(
    dataframe = data_train,
    directory = "./",
    x_col = "path",
    y_col = "action",
    target_size = size,
    class_mode = "categorical",
    batch_size = batch_size,
    seed = 1
)

test_set = data_generator.flow_from_dataframe(
    dataframe = data_test,
    directory = "./",
    x_col = "path",
    y_col = "action",
    target_size = size,
    class_mode = "categorical",
    batch_size = batch_size,
    seed = 1,
    shuffle = False
)

val_set = data_generator.flow_from_dataframe(
    dataframe = data_val,
    directory = "./",
    x_col = "path",
    y_col = "action",
    target_size = size,
    class_mode = "categorical",
    batch_size = batch_size,
    seed = 1
)
````

### Load model to use as feature extractor

What we do here: we use a MobileNet V2 model for FEATURE EXTRACTION. We do this by pre-loading the model and removing the very top (classification layer). The next layer after this will be basically ideal as a feature vector to use in making classifications with a NEW classification layer

### Add classification layer on top of feature extractor

````python
# FROM SCRATCH MODEL WHICH REQUIRES RESCALE

# layer design partly from: https://towardsdatascience.com/cnn-approach-for-predicting-movie-genre-from-posters-95f122f88bc2

base_model = tf.keras.applications.MobileNetV2(input_shape=(224, 224, 3),
                                               include_top=False,
                                               weights='imagenet')


# determine which part is trainable
N_UNFREEZE = 3
for index, layer in enumerate(base_model.layers):
  if index >= len(base_model.layers)-N_UNFREEZE:
    layer.trainable = True
  else:
    layer.trainable = False

# build sequence
from tensorflow.keras.optimizers import RMSprop

model = tf.keras.models.Sequential([
    tf.keras.layers.Rescaling(1./127.5, offset=-1),
    base_model, # needs -1, 1 input pixel values in format 224, 224 3
    tf.keras.layers.SpatialDropout2D(0.25),
    tf.keras.layers.Conv2D(1280,(1,1), padding = 'same'),
    tf.keras.layers.SpatialDropout2D(0.25),
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(2, activation='softmax')
])

# compile
model.compile(loss='categorical_crossentropy',
    optimizer=RMSprop(learning_rate=0.001),
    metrics=['categorical_accuracy'])


model.build(input_shape = (None, 224, 224, 3))

# compile
model.summary()



# # original
#     tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(224, 224, 3)),
#     tf.keras.layers.MaxPooling2D(2,2),
#     tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
#     tf.keras.layers.MaxPooling2D(2,2),
#     tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
#     tf.keras.layers.MaxPooling2D(2,2),
#     tf.keras.layers.Flatten(),
#     tf.keras.layers.Dense(1024, activation='relu'),
#     tf.keras.layers.Dense(2, activation='softmax')
````

````python
correct = 0
total = 0

true0_predict0 = 0
true0_predict1 = 0
true1_predict0 = 0
true1_predict1 = 0

for i in range(33):
  image_batch, label_batch = test_set[i]
  predictions = model.predict(image_batch)
  predictions = [np.argmax(prediction) for prediction in predictions]
  labels = [np.argmax(label) for label in label_batch]
  correct += sum([prediction == label for prediction, label in zip(predictions, labels)])
  total += len(predictions)

  true0_predict0 += sum([prediction == 0 and label == 0 for prediction, label in zip(predictions, labels)])
  true0_predict1 += sum([prediction == 1 and label == 0 for prediction, label in zip(predictions, labels)])
  true1_predict0 += sum([prediction == 0 and label == 1 for prediction, label in zip(predictions, labels)])
  true1_predict1 += sum([prediction == 1 and label == 1 for prediction, label in zip(predictions, labels)])


print(correct)
print(total)
print("Accuracy", correct/total)

print("T0,P0", true0_predict0)
print("T0,P1", true0_predict1)
print("T1,P0", true1_predict0)
print("T1,P1", true1_predict1)
````
